# JPL Flight Data
    Azimuth and elevation angles given in degrees
    Resolution of azimuth and elevation data appears to be ~0.6 degrees
    Looks like coarse pointing system kept balloon +/- 12 degrees in azimuth
    
    

```python
%matplotlib inline
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
################################################################################
base_size=15
mpl.rcParams['legend.fontsize'] = base_size
mpl.rcParams['figure.figsize'] = (10,7.5)
mpl.rcParams['figure.titlesize']=base_size+5
mpl.rcParams['xtick.labelsize']=base_size
mpl.rcParams['ytick.labelsize']=base_size
mpl.rcParams['font.size']=base_size
mpl.rcParams['axes.titlesize']=base_size
mpl.rcParams['axes.labelsize']=base_size
mpl.rcParams['lines.markersize'] = 5           # markersize, in points
mpl.rcParams['legend.markerscale'] = 1     # line width in points
mpl.rcParams['lines.markeredgewidth'] = 0.2 # the line width around the marker symbol
mpl.rcParams['lines.linewidth'] = 1.5
#####################################
```

```python
file_loc = "C:/git_repos/GLO/tracking/sample_balloon_data/JPL_flight/eng16271.out"
data = pd.read_csv(file_loc,header=2,delim_whitespace=True)
data = data.set_index('tsecs')

mask_1 = (data.index > 62558) & (data.index < 90036)
mask_2 = (data.index > 62000) & (data.index < 70000)
mask_3 = (data.index > 63779) & (data.index < 63796)
```

```python
plt.plot(data.loc[:,'AZIMUTH_Angle'],label='Azimuth')
plt.plot(data.loc[:,'ELEVATION_Angle'],label='Elevation')
plt.xlabel('Time Elapsed (seconds)')
plt.ylabel('Degrees')
plt.title('JPL Flight Data: Entire Flight')
plt.legend()
```

```python
plt.plot(data.loc[mask_1,'AZIMUTH_Angle'],label='Azimuth')
plt.xlabel('Time Elapsed (seconds)')
plt.ylabel('Degrees')
plt.title('JPL Flight Data')
plt.legend()
```

```python
plt.plot(data.loc[mask_2,'AZIMUTH_Angle'],label='Azimuth')
plt.plot(data.loc[mask_3,'AZIMUTH_Angle'],color='r',label='Steep Slope\napprox 0.65 deg/sec')
plt.xlabel('Time Elapsed (seconds)')
plt.ylabel('Degrees')
plt.title('JPL Flight Data')
plt.legend()
```

```python
plt.plot(data.loc[mask_3,'AZIMUTH_Angle'],color='r',label='Azimuth\nslope approx 0.65 deg/sec')
plt.xlabel('Time Elapsed (seconds)')
plt.ylabel('Degrees')
plt.title('JPL Flight Data: Zoomed in on a "Steep" slope')
plt.legend()
```
