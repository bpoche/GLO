4.6 - Output Filter (values 0.0 to 1.0)
This filter can reduce jittery movements and make controls smoother. Increasing filter strength (lower 
values) lengthens the life of the PTU but can negative
ly impact system performance at high disturbance 
frequency

set pan axis filter to 0.5
SCPKS.5*

The following example sets the stabilization pan axis proportional gain and then verifies the setting:
SCPKP16.00000 *
SCPKP * 16.00000

4.7 - Adjust Controller Acceleration
Feed-Forward
Query or set the pan and tilt acceleration feed-for
ward. This determines how far ahead in samples the 
PTU attempts to estimate. If the PTU undergoes reason
ably smooth motion (such as a boat or a vehicle 
with good suspension), leave this value at defa
ult. Decrease this value for rougher movements
